```python
import pandas as pd
```

```python
titanic1 = pd.read_csv('train.csv')
```

```python
titanic1.head()
```

```python
titanic1.Sex=titanic1.Sex.map({'male':0,'female':1})
```

```python
ports=pd.get_dummies(titanic1.Embarked, prefix = 'Embarked')
```

```python
titanic1=titanic1.join(ports)
```

```python
titanic1.drop(['Embarked'],axis=1, inplace=True)
```

```python
y=titanic1.Survived.copy()
x=titanic1.drop(['Survived'],axis=1)
```

```python
x.drop(['Name', 'Cabin', 'Ticket', 'PassengerId'], axis = 1, inplace = True)
```

```python
x.info()
```

```python
x.isnull().values.any()
```

```python
x[pd.isnull(x).any(axis=1)]
```

```python
x.Age.fillna(x.Age.mean(), inplace= True)
```

```python
x.isnull().values.any()
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_valid, y_train, y_valid =train_test_split(x,y,test_size=0.2, random_state=7)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_train, y_train)
```

```python
model.score(x_valid, y_valid)
```
